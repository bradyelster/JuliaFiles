```julia
using Plots
using LaTeXStrings
using Printf
```

```julia
x = collect(LinRange(0, 1, 1000))
ϵ = 0.01
α = 1
β = 2
A = β - α + 1
B(x, ϵ) = 2x*(1 - (1/ϵ)) + 3

inner1(x, ϵ) = β + A*(exp((x-1)/ϵ) - 1)
inner2(x, ϵ) = B(x, ϵ)*(exp((x-1)/ϵ) - 1) - 2*((x-1)/ϵ)

outer1(x) = α - x^2
outer2(x) = -2*x;
```

```julia
INNER(x, ϵ) = inner1.(x,ϵ) .+ ϵ*inner2.(x, ϵ)
outer1(x) = α - x^2
outer2(x) = -2*ϵ*x
OUTER(x, ϵ) = outer1.(x) .+ outer2.(x)
MATCH(x, ϵ) = α - 2*ϵ*x
COMP(x, ϵ) = INNER(x, ϵ) .+ OUTER.(x, ϵ) .- MATCH.(x, ϵ)
```

```julia
# Initialize an empty plot
theme(:dao::Symbol)
plt = plot(x, [inner1.(x, ϵ), outer1.(x)], 
    xlabel = L"$x$",  # Explicitly set the x-axis label
    ylabel = L"$y(x)$",
    title = "Problem 2 Solution "*L"$(\varepsilon=%$ϵ)$",
    legend = :best,
    titlefontsize = 20,
    tickfontsize = 12,
    legendfontsize = 10,
    yguidefontsize = 15,
    xguidefontsize = 15,
    line=(3, :solid),
    # right_margin = 2 * Plots.mm
)

plt
```
