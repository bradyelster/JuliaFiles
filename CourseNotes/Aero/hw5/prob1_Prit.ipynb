$\frac{1}{6}\left(-3+\Omega^2+2 \Omega^2 \operatorname{Cos}[t \Omega]\right) \operatorname{Sin}\left[\frac{t \Omega}{2}\right]^2$

```julia
using OrdinaryDiffEq
using Plots
```

```julia
A = 0.1;
ω = π # pi rad/s
p = 1
g = 10 
Ω = g/(ω^2*p);
```

```julia
x₁(t) = 1/6(-3+Ω^2+2*Ω^2*cos(Ω*t))*sin(Ω*t/2)^2;
tvals = LinRange(0, 15, 500);
```

```julia
plot(tvals, x₁.(tvals),
    line=(3, :solid),
    )
```
