```julia
using CairoMakie, LaTeXStrings, SpecialFunctions
```

```julia
# create array of gridpoints
x = 0:0.1:15

# create figure: specify size and fonts
fig = Figure(
    size = (600, 400), 
    fonts = (; regular = "CMU Serif")
)

# initialize empty plot: specify axes labels, grid, tick marks
ax = Axis(
    fig[1, 1], 
    xlabel = L"x", 
    ylabel = L"J_{\nu}(x)",
    ylabelsize = 22,
    xlabelsize = 22,
    xgridstyle = :dash,
    ygridstyle = :dot,
    xtickalign = 0, # tick marks facing out
    ytickalign = 1, # tick marks facing in
    xticksize = 10, 
    yticksize = 10, 
    xlabelpadding = -10 # brings x-axis label closer to x-axis
)

# make array of linestyles for each Bessel function
styles = [:solid, :dash, :dot, :dashdot, :dash, :dot, :dashdot]

# add Bessel functions to the plot
for (i, ) in enumerate(0:6)
    lines!(ax, x, besselj.(, x), 
           label = latexstring("J_{$()}(x)"),
           linestyle = styles[i],
           linewidth = 2)
end

# create legend
axislegend(; nbanks = 2, framecolor = (:grey, 0.5));

# display figure
fig
```
