```julia
using Plots
using LaTeXStrings
using Printf
gr()
theme(:dao)
# pythonplot() - not as good imo
```

```julia
# Define constants
VA = 1.0  # Alfvén speed (normalized)
θ = range(0, 2π, length=1000);  # Angle θ from 0 to 2π
```

```julia
# Function to calculate phase speeds for the three modes
function phase_speeds(cs2_VA2)
    cs = sqrt(cs2_VA2) * VA  # Sound speed
    Vp_alfven = VA .* abs.(cos.(θ))  # Shear Alfvén mode
    Vp_fast = @. sqrt(0.5 * ((VA^2 + cs^2) + sqrt((VA^2 + cs^2)^2 - 4 * VA^2 * cs^2 * cos(θ)^2)))  # Fast mode
    Vp_slow = @. sqrt(0.5 * ((VA^2 + cs^2) - sqrt((VA^2 + cs^2)^2 - 4 * VA^2 * cs^2 * cos(θ)^2)))  # Slow mode
    return Vp_alfven, Vp_fast, Vp_slow
end
```

```julia
# Plotting function
function plot_WNS(cs2_VA2, title)
    Vp_alfven, Vp_fast, Vp_slow = phase_speeds(cs2_VA2)
    plot(θ, Vp_alfven, proj=:polar, label="Shear Alfvén", linewidth=2, legend=:best)
    # plot!(θ, Vp_fast, proj=:polar, label="Fast Magnetosonic", linewidth=2)
    plot!(θ, Vp_slow, proj=:polar, label="Slow Magnetosonic", linewidth=2,  ls=:dash,
    dpi=300,
    #size=(400,400),
    left_margin=-10* Plots.mm,
    right_margin=-1* Plots.mm,
    titlefontsize=15,
    tickfontsize=10,
    legendfontsize=8,
    yguidefontsize=15,
    xguidefontsize=15)
    title!(title)
end
```

```julia
# Generate plots for the three cases
cs2_VA2_values = 2000
titles = "blah"
```

```julia
plot_WNS(cs2_VA2_values, titles)
```
